# Tâche 7 : étude des pics du diffractogramme pour le carbone (mp-990448) 

## Imports

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

```python
with MPRester("GfTnOwQgZddh8BopXvNa") as m:
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-990448") 
```

## Affichage des trois premiers pics

```python
xrdcalc = XRDCalculator(wavelength = 'CuKa', symprec = 0, debye_waller_factors = None)
xrdcalc.show_plot(structure,two_theta_range=(0,35),annotate_peaks=True,with_labels=True)
```

```python
xrdpattern = xrdcalc.get_pattern(structure = structure, scaled = True, two_theta_range = (0,90))
```

## Premier pic

```python
angle1 = xrdpattern.x[0] ; intensite1 = xrdpattern.y[0]
hkl1 = xrdpattern.hkls[0][0]['hkl'] ; multi1 = xrdpattern.hkls[0][0]['multiplicity']
#print("l'angle 2 thétas = {}; l'intensité du pic = {}; [hkl] = {}; la multiplicité = {}.".format(angle1,intensite1,hkl1,multi1))
```

Le premier pic possède les caractéristiques (pouvant se retrouver sur le graphe ci-dessus) suivantes. 
L'angle 2$\theta$ est égal à 8.84370111670552$°$.$\newline$
L'intensité du pic est de 100.$\newline$
Les indices hkl associés sont (0, 0, 0, 1).$\newline$
La multiplicité est de 2.$\newline$
